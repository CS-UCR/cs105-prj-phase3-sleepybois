```python
# Import libraries
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
from numpy import * # everything 
import random as rand
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab



# Create dataframes
df_ca = pd.read_csv("./Videos/CAvideos.csv")
df_de = pd.read_csv("./Videos/DEvideos.csv")
df_fr = pd.read_csv("./Videos/FRvideos.csv")
df_gb = pd.read_csv("./Videos/GBvideos.csv")
df_in = pd.read_csv("./Videos/INvideos.csv")
df_jp = pd.read_csv("./Videos/JPvideos.csv", encoding="ISO-8859-1")
df_kr = pd.read_csv("./Videos/KRvideos.csv", encoding="ISO-8859-1")
df_mx = pd.read_csv("./Videos/MXvideos.csv", encoding="ISO-8859-1")
df_ru = pd.read_csv("./Videos/RUvideos.csv", encoding="ISO-8859-1")
df_us = pd.read_csv("./Videos/USvideos.csv")
```

> The following EDA examines some of the Quantitative Variables from Youtube Videos from different regions. Views, Comments, Likes, and Dislikes from USA, Great Britain, and Canada were examined. Data cleaning was also peformed to remove duplicates

## Data Cleaning 

```python
df_us.info()
```

### Remove Duplicates

```python
print('# of US rows before cleaning duplicates: ', df_us.shape[0])
df_us.drop_duplicates(keep="first", inplace=True)
print('# of US rows after cleaning duplicates: ', df_us.shape[0], '\n')
print('# of GB rows before cleaning duplicates: ', df_gb.shape[0])
df_gb.drop_duplicates(keep="first", inplace=True)
print('# of GB rows after cleaning duplicates: ', df_gb.shape[0], '\n')
print('# of CA rows before cleaning duplicates: ', df_ca.shape[0])
df_ca.drop_duplicates(keep="first", inplace=True)
print('# of CA rows after cleaning duplicates: ', df_ca.shape[0], '\n')
```

# Distribution of Likes/Dislikes/Views for each region

## USA

```python
fig=plt.figure()
plt.scatter(df_us.likes, df_us.dislikes)
axis = fig.gca() #get current axis
axis.set_title('USA : Likes vs Dislikes')
axis.set_xlabel('Likes')
axis.set_ylabel('Dislikes')
fig.canvas.draw()

fig=plt.figure()
plt.scatter(df_us.likes, df_us.views)
axis = fig.gca() #get current axis
axis.set_title('USA : Likes vs Views')
axis.set_xlabel('Likes')
axis.set_ylabel('Views')
fig.canvas.draw()

fig=plt.figure()
plt.scatter(df_us.dislikes, df_us.views)
axis = fig.gca() #get current axis
axis.set_title('USA : Dislikes vs Views')
axis.set_xlabel('Dislikes')
axis.set_ylabel('Views')
fig.canvas.draw()
```

> For the US Videos, there seems to be a positive correlation between likes and views and likes and dislikes and views. Other than a couple of outliers, the scatter plots suggest that more views will lead to more likes and dislikes on Youtube Videos from the US. The other regions have scatter plots that display a similar relationship. 

```python
n, bins, patches = plt.hist(df_us['likes'], 25, facecolor='green')
plt.title('USA : Histogram of Likes')
plt.show()
n, bins, patches = plt.hist(df_us['dislikes'], 25, facecolor='green')
plt.title('USA : Histogram of Dislikes')
plt.show()
n, bins, patches = plt.hist(df_us['views'], 25, facecolor='green')
plt.title('USA: Histogram of Views')
plt.show()
n, bins, patches = plt.hist(df_us['comment_count'], 25, facecolor='green')
plt.title('USA: Histogram of Comments')
plt.show()
```

> The above distributions are positvely skewed for views, comments, likes and dislikes. These disributions look similar for other regions like Great Britain and Canada.

## Statistical Summary

### Totals

```python
print("US Likes: " + str(df_us['likes'].sum()))
print("US views: " + str(df_us['views'].sum()))
print("GB Likes: " + str(df_gb['likes'].sum()))
print("GB views: " + str(df_gb['views'].sum()))
print("CA Likes: " + str(df_ca['likes'].sum()))
print("CA views: " + str(df_ca['views'].sum()))
```

```python

print("US Like to views ratio : " + str(df_us['likes'].sum() / df_us['views'].sum()))
print("GB Like to views ratio : " + str(df_gb['likes'].sum() / df_gb['views'].sum()))
print("CA Like to views ratio : " + str(df_ca['likes'].sum() / df_ca['views'].sum()))

```

> By calculating the total amount of likes and views of all the videos from different region, we can calculate the like to view ratio, which is one way to determine which region has the more popular videos. It seems as if videos from the Canadaian region are the most popular out of Great Britain and the US. 

### Averages

```python
print("US Likes: " + str(round(df_us['likes'].mean(), 0)))
print("US views: " + str(round(df_us['views'].mean(), 0)))
print("GB Likes: " + str(round(df_gb['likes'].mean(), 0)))
print("GB views: " + str(round(df_gb['views'].mean(), 0)))
print("CA Likes: " + str(round(df_ca['likes'].mean(), 0)))
print("CA views: " + str(round(df_ca['views'].mean(), 0)))
```

```python
print("US Like to views ratio : " + str(round(df_us['likes'].mean(), 0) / round(df_us['views'].mean(), 0)))
print("GB Like to views ratio : " + str(round(df_gb['likes'].mean(), 0) / round(df_gb['views'].mean(), 0)))
print("CA Like to views ratio : " + str(round(df_ca['likes'].mean(), 0) / round(df_ca['views'].mean(), 0)))
```
